## Correlation
For this part, I decided to analyze New York counties with the social, economic, and housing enrichment dataset.

```python
ny_cases = cases.loc[cases['State'] == 'NY']
ny_cases = ny_cases.drop(ny_cases.columns[:4], axis=1)
ny_cases = ny_cases.iloc[1: , :]
ny_cases.index += 2
ny_cases.head()
```

Calculating new cases

```python
ny_new_cases = pd.DataFrame()
for column in range(1, 573):
  ny_new_cases[column] = ny_cases.iloc[:,column] - ny_cases.iloc[:,column - 1]
ny_new_cases.head()
```

Calculating the mean and normalizing by 10000

```python
ny_avg_cases = ny_new_cases.mean(axis=1)
ny_population = population.loc[population['State'] == 'NY']
ny_population = ny_population.iloc[2: , :]
ny_normalized_avg_cases = (ny_avg_cases / ny_population['population']) * 10000
ny_normalized_avg_cases = ny_normalized_avg_cases.round()
ny_normalized_avg_cases.head()
```

Merging cases with FIPS code

```python
ny_counties = cases.loc[cases['State'] == 'NY']
ny_counties = ny_counties.iloc[1: , :]
ny_counties.index += 2

ny = pd.DataFrame([ny_counties['countyFIPS'], ny_normalized_avg_cases])
ny = ny.T
ny.head()
```

Merging enrichment set with New York county cases

```python
ny_enrichment = soc_econ_house_cases[soc_econ_house_cases['State_cases'] == 'NY']
ny_merged = ny_enrichment.merge(ny, left_on='countyFIPS_cases', right_on='countyFIPS')
ny_merged = ny_merged.drop(ny_merged.columns[:1095], axis=1)
ny_merged = ny_merged.rename(columns={'Unnamed 0': 'Normalized_Cases'})
ny_merged.head()
```

Calculating correlation between New York cases and number of married-couple families

```python
X = ny_merged['Normalized_Cases']
Y = pd.to_numeric(ny_merged.iloc[:, 1].replace(',','', regex=True))
np.corrcoef(X, Y)[0, 1]
```

Calculating correlation between New York cases and number of householders living alone

```python
Y = pd.to_numeric(ny_merged.iloc[:, 7].replace(',','', regex=True))
np.corrcoef(X, Y)[0, 1]
```
